```python
import random
population = []
#################################################################################
def fitness(seq):
    boardsize=8
    score = 0
    for row in range(boardsize):
        col = seq[row]
        for row1 in range(boardsize):
            if row1 == row:
                continue
            if seq[row1] == col:
                continue
            if row1 + seq[row1] == row + col:
                continue
            if row1 - seq[row1] == row - col:
                continue
            score += 1
    return score / 2
#################################################################################
def fsort():
    for i in range(len(population)):
        for j in range(i + 1, len(population)):
            if fitness(population[i]) < fitness(population[j]):
                population[i], population[j] = population[j], population[i]
#################################################################################
def selector(percent):
    selected = []
    total=0
    for i in range(len(population)):
        total += fitness(population[i])
    # Percent to number(int)
    n = len(population)
    n=int(n/100)
    n=n*percent

    r = 0
    # selects num number of individual
    for i in range(n):
        r = random.randint(0, total)
        #ind = rangeInd(population,r)
        total1 = 0
        for i1 in range(len(population)):
            total1+=fitness(population[i1])
            if total1>=r:
                ind=i1
                break
        total-=fitness(population[ind])
        selected.append(population[ind])
        list.pop(ind)
        
    return selected
#################################################################################
def cross(c1, c2):
    crossPoint1=random.randint(0,len(c1)-1)
    crossPoint2=random.randint(0,len(c2)-1)
    index=0
    for x in range(crossPoint1,crossPoint2):
        temp=c2[index]
        c2[index]=c1[x]
        c1[x]=temp
        index+=1
    return c1,c2
#################################################################################
def crossOver():
    crossed=[]
    n = int(len(population)/3)
    j = 1
    if (n % 2) != 0:
        n -= 1
    for i in range(n):
        c1=population[0]
        population.pop(0)
        if (n-(i+j)) > 0:
            r=random.randint(0, (n-(i+j)))
        else:
            r=0
        c2 = population[r]
        population.pop(r)
        ret=cross(c1,c2)
        crossed.append(ret[0])
        crossed.append(ret[1])
        j += 1
    return crossed
#################################################################################
def mutation():
    n=len(population)
    mutated=[]
    for i in range(n):
        x=population[len(population) - 1 - i]
        x[random.randint(0, len(x) - 1)] = random.randint(1, 8)
        x[random.randint(0, len(x) - 1)] = random.randint(1, 8)
        ind=x
        population.pop(i)
        population.append(ind)
    return mutated
#################################################################################
size=20     #size of papulation
boardsize=8;#no. of queens and bord dimentions
#Genrating Papulation
for i in range(size):
        temp = []
        for j in range(boardsize):
            temp.append(random.randint(1, boardsize))
        population.append(temp)
fsort()
print("Max Fitness Of new Population : ",fitness(population[0]))
print("Max Fit Indiviual : ",population[0])
print("\n -: After Appling the Genetic Algo :-\n")
i = 0
while i < 500 and fitness(population[0])<26:
    selected = selector(50)
    crossed = crossOver()
    mutated = mutation()
    population.extend(selected)
    population.extend(crossed)
    population.extend(mutated)
    fsort()
    i += 1
print("Max Fitness Of Genetically Modified Population : ",fitness(population[0]))
print("Max Fit Modified Indiviual : ",population[0])
print("Genration Number # ",i)
```
